```python
###Removing outliers
```

```python
df1.describe()
```

```python
# removing outlier for FARE
```

```python
min_fare ,max_fare = df1.Fare.quantile([.001,.999])
min_fare ,max_fare
```

```python
df1[df1.Fare < min_fare]
```

```python
df1[df1.Fare > max_fare]
```

```python
df2 = df1[(df1.Fare > min_fare) & (df1.Fare < max_fare)]
df2
```

```python
# removing outliers for AGE
```

```python
min_age ,max_age = df2.Age.quantile([.001,.999])
min_age ,max_age
```

```python
df2[df2.Age < min_age]
```

```python
df2[df2.Age > max_age]
```

```python
df3 = df2[(df2.Age > min_age) & (df2.Age < max_age)]
df3
```
